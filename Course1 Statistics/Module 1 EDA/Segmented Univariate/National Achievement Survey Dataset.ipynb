```python
num_cols=['Maths..', 'Reading..', 'Science..', 'Social..']
cat_cols=["Gender","Age","Siblings","Handicap",'Father.edu', 'Mother.edu',
       'Father.occupation', 'Mother.occupation', 'Below.poverty','Watch.TV',
       'Read.magazine', 'Read.a.book', 'Play.games', 'Help.in.household']
```

```python
for i in num_cols:
    print("Boxplot of",i)
    sns.boxplot(x=df[i])
    plt.show()
```

```python
for i in num_cols:
    print("Histplot of",i)
    sns.histplot(x=df[i])
    plt.show()
```

```python
for col1 in cat_cols:
    for col2 in num_cols:
        print("Boxplot of",col1,"Vs",col2)
        sns.boxplot(x=df[col1],y=df[col2])
        plt.show()
```
